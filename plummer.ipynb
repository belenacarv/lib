```python
import numpy as np
import matplotlib.pyplot as plt

def potential_plummer(radius, scale_radius, total_mass):
    G = 6.67430e-11  # Constante de gravitación universal
    numerator = -G * total_mass
    denominator = np.sqrt(radius**2 + scale_radius**2)**3
    potential = numerator / denominator
    return potential

# Parámetros de la esfera de Plummer
scale_radius = 1.0  # Radio de escala
total_mass = 1.0  # Masa total

# Generar datos para el gráfico
radii = np.linspace(0.01, 10, 100)  # Valores de radio
potentials = potential_plummer(radii, scale_radius, total_mass)

# Graficar el potencial gravitatorio en función del radio
plt.plot(radii, potentials)
plt.xlabel('Radio')
plt.ylabel('Potencial Gravitatorio')
plt.title('Potencial Gravitatorio de la Esfera de Plummer')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plummer_density(radius, scale_radius, total_mass):
    # Calcula la densidad de la esfera de Plummer en un radio dado
    num = 3.0 * total_mass
    denom = 4.0 * np.pi * scale_radius**3 * (radius**2 + scale_radius**2)**(5/2)
    return num / denom

# Parámetros
scale_radius = 1.0  # Radio característico de la esfera de Plummer
total_mass = 1.0  # Masa total de la esfera de Plummer
num_points = 100  # Número de puntos para graficar

# Rango de valores de radio
radius_values = np.linspace(0.01, 10.0, num_points)

# Calcula la densidad para cada valor de radio
density_values = plummer_density(radius_values, scale_radius, total_mass)

# Graficar la densidad en función del radio
plt.plot(radius_values, density_values)
plt.xlabel('Radio')
plt.ylabel('Densidad')
plt.title('Densidad de la Esfera de Plummer')
plt.grid(True)
plt.show()
```
